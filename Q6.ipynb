```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def scatter_plot_from_file(filename, ax):
    # Load the data using pandas
    data = pd.read_csv(filename, delimiter=" ", header=None)  # assuming the data is space-separated
    
    # Extracting features and labels
    x_values = data[0]
    y_values = data[1]
    labels = data[2]
    
    # Plotting data points with different colors based on label
    ax.scatter(x_values[labels == 0], y_values[labels == 0], color='red', label='Label 0')
    ax.scatter(x_values[labels == 1], y_values[labels == 1], color='blue', label='Label 1')
    
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")
    ax.set_title(f"Scatter plot for {filename}")
    ax.legend()
```

```python
# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))  # Adjust figsize as needed

# Plotting for D1.txt in the first subplot
scatter_plot_from_file("D1.txt", ax1)

# Plotting for D2.txt in the second subplot
scatter_plot_from_file("D2.txt", ax2)

# Adjust spacing between subplots if necessary
plt.tight_layout()

# Show the figure with both subplots
plt.show()
```
